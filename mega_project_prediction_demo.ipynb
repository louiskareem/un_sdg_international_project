#### Import the essential packages

```python
from keras.models import load_model
from keras.preprocessing import image
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import sys, os

model = load_model('classifier_three.h5')
%matplotlib inline 
```

#### Prediction 

```python
def prediction(val):
    path = 'organic_and_recyclable/'
    for f in os.listdir("organic_and_recyclable"):
        # load an image to be tested (random image from Google)
        img_pred = image.load_img('organic_and_recyclable/'+val+'.jpg', target_size=(200,200))

        # convert image to numpy array
        img_pred = image.img_to_array(img_pred)

        # expand the array
        img_pred = np.expand_dims(img_pred, axis=0)

        # predict the image from the img_pred
        result = model.predict(img_pred)

        # # round the result to a more readable number
        results = np.round(result)

        if result[0][0] == 1 and f == val+'.jpg':
            prediction = 'Recyclable'
            img = mpimg.imread('organic_and_recyclable/'+f)
            plt.imshow(img, interpolation='nearest')
            plt.axis('off')
            plt.title(prediction)
            plt.show()
            break
        elif result[0][0] == 0 and f == val+'.jpg':
            prediction = 'Organic'
            img = mpimg.imread('organic_and_recyclable/'+f)
            plt.imshow(img, interpolation='nearest')
            plt.axis('off')
            plt.title(prediction)
            plt.show()
            break
```

```python
val = input()

if __name__ == '__main__':
    prediction(val)
```
